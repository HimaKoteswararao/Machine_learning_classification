```python
#initialising independent,target data from dataset
ind_data = df.select_dtypes(include=['int64', 'float64']).columns
print(ind_data)
target_data = df.select_dtypes(include=['object']).columns
print(target_data)
```

```python
#plotting bar graph count for class
for data in target_data:
    plt.figure(figsize=(10,5))    
    sns.barplot(
        x = df[data].value_counts(), 
        y = df[data].value_counts().index, 
        data = df)  
    for i, count in enumerate(df[data].value_counts()):
        plt.text(count, i, str(count), ha='left', va='center')
    plt.title('Bar graph plotting count of each class')    
    plt.tight_layout()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Class']=le.fit_transform(df['Class'])
```

```python
dict_le = dict(zip(le.transform(le.classes_),le.classes_))
print(dict_le)
```

```python
df.head(10)
```

```python
# Separate the features and target variable
X = df.drop('Class', axis=1)
Y = df['Class']
```

```python
from sklearn.preprocessing import MinMaxScaler

# Assuming you have a DataFrame called 'df' with the dataset

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Perform Min-Max scaling on the dataset
df_normalized = scaler.fit_transform(df)

# Convert the scaled array back to a DataFrame
df_normalized = pd.DataFrame(df_normalized, columns=df.columns)
df_normalized.describe()
```

```python
#splitting of data for train and test
from sklearn.model_selection import train_test_split

# Split the data into training and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

# SVM

```python
from sklearn.metrics import accuracy_score

from sklearn.svm import SVC
# Train an SVM model
svm = SVC()  
svm.fit(X_train, Y_train)

# Predict on the train set and calculate accuracy 
y_pred = svm.predict(X_train)  
accuracy = svm.score(X_train, Y_train)
print(accuracy)
```

```python
# Train an SVM model
svm = SVC()  
svm.fit(X_train, Y_train)

# Predict on the test set and calculate accuracy 
y_pred = svm.predict(X_test)  
accuracy = svm.score(X_test, Y_test)
print(accuracy)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.exceptions import ConvergenceWarning
import warnings

# Ignore the ConvergenceWarning
warnings.filterwarnings("ignore", category = ConvergenceWarning)

# Train a logistic regression model
logreg = LogisticRegression(solver='lbfgs', max_iter=1000)
logreg.fit(X_train, Y_train)

# Predict on the test set
y_pred = logreg.predict(X_train)

# Calculate accuracy
accuracy = logreg.score(X_train, Y_train)
print(accuracy)
```

```python
# Train a logistic regression model
logreg = LogisticRegression(solver='lbfgs', max_iter=1000)
logreg.fit(X_train, Y_train)

# Predict on the test set
y_pred = logreg.predict(X_test)

# Calculate accuracy
accuracy = logreg.score(X_test, Y_test)
print(accuracy)
```

# Naive Bayesian

```python
from sklearn.naive_bayes import GaussianNB

GaussNB = GaussianNB()
GaussNB.fit(X_train, Y_train)
y_pred = GaussNB.predict(X_train)
accuracy = GaussNB.score(X_train, Y_train)
print(accuracy)
```

```python
GaussNB = GaussianNB()
GaussNB.fit(X_test, Y_test)
y_pred = GaussNB.predict(X_test)
accuracy = GaussNB.score(X_test, Y_test)
print(accuracy)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators=500, max_depth=10)
forest.fit(X_train, Y_train)
y_pred = forest.predict(X_train)
accuracy = forest.score(X_train, Y_train)
print(accuracy)
```

```python
forest = RandomForestClassifier(n_estimators=500, max_depth=10)
forest.fit(X_train, Y_train)
y_pred = forest.predict(X_test)
accuracy = forest.score(X_test, Y_test)
print(accuracy)
```
